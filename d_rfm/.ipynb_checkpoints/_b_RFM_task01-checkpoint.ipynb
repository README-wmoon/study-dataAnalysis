### RFM 분석 실습

- Recency: 얼마나 최근에 구매했는가
- Frequency: 얼마나 자주 구매했는가
- Monetary: 얼마나 많은 금액을 지출했는가

### 고객 분석
- CustomerID: 고객 번호
- PurchaseDate: 구매 날짜
- TransactionAmount: 결제 금액 
- ProductInformation: 상품 정보
- OrderID: 주문 번호
- Location: 주문 지역

```python
import pandas as pd
path = './datasets/customer_order_data.csv'


# parse_dates 날짜형식으로 되어있는 데이터를 datetime형식으로 바꿔주는 함수
co_df = pd.read_csv(path, parse_dates=['PurchaseDate'])
display(co_df)
```

```python
co_df.info()
```

```python
co_df.duplicated().sum()
```

```python
co_df.isnull().sum()
```

```python
co_df.astype({'PurchaseDate': 'datetime64[ns]'})
```

```python
co_df[co_df["PurchaseDate"] < '2023']
```

```python
import numpy as np

Q1 = co_df.describe().T.loc['TransactionAmount', '25%']
Q3 = co_df.describe().T.loc['TransactionAmount', '75%']

iqr = Q3 - Q1

lower_bound = Q1 - iqr * 1.5
upper_bound = Q3 + iqr * 1.5

if lower_bound < 0:
    lower_bound = 0


print(f'정상치 범위: {lower_bound} ~ {upper_bound}')
```

```python
co_df = co_df.sort_values(['PurchaseDate'], ascending=False)
co_df
```

```python
co_df.info()
```

```python
co_df['CustomerID'].value_counts()
```

```python
from datetime import datetime

# 현재 날짜 가져오기
current_date = datetime.now()

# 'PurchaseDate' 열의 값과 현재 날짜의 차이를 일(day) 수로 계산하여 'Year_Customer' 열에 할당
co_df['Year_Customer'] = (current_date - co_df['PurchaseDate']).dt.days

# 결과 확인
display(co_df)
```

```python
co_df['Recency'] = co_df.groupby('CustomerID')['Year_Customer'].transform('max')
display(co_df)
```

```python
co_df['Frequency'] = co_df.groupby('CustomerID')['CustomerID'].transform('count')

display(co_df)
```

```python
co_df = co_df.drop(labels=['Year_Customer'], axis=1)
co_df
```

```python
co_df['Monetary'] = co_df.groupby('CustomerID')['TransactionAmount'].transform('sum')
display(co_df)
```

```python
co_df[co_df['CustomerID'] == 4804]
```

```python
co_df['ProductInformation'] = co_df['ProductInformation'].replace(
    ['Product A', 'Product B', 'Product C', 'Product D'], ['Mouse', 'Computer', 'Monitor', 'Keyboard']
)

co_df
```

```python
co_rfm_df = co_df[['Recency', 'Frequency', 'Monetary']]
co_rfm_df
```

```python
co_df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler

normalization = MinMaxScaler()
rfm_normalization = normalization.fit_transform(co_rfm_df)
```

```python
co_rfm_df = pd.DataFrame(rfm_normalization, columns=co_rfm_df.columns)
co_rfm_df
```

```python
co_df[co_df['CustomerID'] == 4804]
```

```python
co_df['TotalScore'] = co_rfm_df['Recency'] + co_rfm_df['Frequency'] + co_rfm_df['Monetary']
```

```python
co_df = co_df.drop(columns=['Recency', 'Frequency', 'Monetary'], axis=1)
```

```python
co_df
```

```python
co_df[co_df['CustomerID'] == 4804]
```

```python
import numpy as np

l1, l2, l3, l4 = np.percentile(co_df['TotalScore'],[20, 50, 70, 90])
print(l1, l2, l3, l4)
```

```python
def get_level(x):
    if x <= l1:
        return 5
    if x > l1 and x <= l2:
        return 4
    if x > l2 and x <= l3:
        return 3
    if x > l3 and x <= l4:
        return 2
    return 1
```

```python
co_df['Level'] = co_df['TotalScore'].apply(get_level)
```

```python
co_df
```

```python
co_df['Level'] = co_df['Level'].replace([5, 4, 3, 2, 1], ['Bronze', 'Silver', 'Gold', 'Diamond', 'VIP'])
```

```python
co_df
```

```python
print(co_df['Level'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

order_columns = ['Bronze', 'Silver', 'Gold', 'Diamond', 'VIP']

sns.countplot(x='Level', data=co_df, palette='muted', order=order_columns)
```

```python
co_df
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

order_columns = ['Bronze', 'Silver', 'Gold', 'Diamond', 'VIP']

# 어떤 고객 ID가 어떤 레벨인지 알고 싶었습니다.
sns.boxplot(ax=axes[0], x='Level', y='CustomerID', data=co_df, palette='muted', order=order_columns)

# 레벨마다 결제 금액은 어느정도 일지 궁금하다
sns.boxplot(ax=axes[1], x='Level', y='TransactionAmount', data=co_df, palette='muted', order=order_columns)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

order_columns = ['Bronze', 'Silver', 'Gold', 'Diamond', 'VIP']

sns.countplot(ax=axes[0], x='ProductInformation', hue='Level', data=co_df, palette='muted', hue_order=order_columns)
sns.countplot(ax=axes[1], x='Location', hue='Level', data=co_df, palette='muted', hue_order=order_columns)
```

### 마케팅 전략 제안
- 상품 정보가 낮은 'VIP' 등급의 고객들을 대상으로 상품을 사게 만드는 광고나 프로모션을 하여 고객 이탈을 방지 할 수 있습니다.
- 주문 지역이 대체로 많은 Tokyo를 대상으로 상품을 홍보하거나 서비스를 제공하는 것이 효과적일 수가 있다.
- 'Silver' 등급의 고객들이 상품구매 수요가 높아 대대적으로 마케팅을 하는 것이 효과적일 수 있습니다.
- 가장 많이 팔린 Monitor의 공급을 높여 소비자들에게 소비를 유도하는 광고나 홍보를 하여 마케팅을 하는 것이 효과적일 수 있습니다.
