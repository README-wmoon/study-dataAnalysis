```python
#describe() 로 데이터 수, 평균, 표준편차, 최소값, 최댓값, 사분위 분포도 확인
# 이상치를 먼저 구하고 제거한다음 결측치를 구하는 거였는데 확인하겠습니다.
display(happiness_df.describe().T)
```

```python
import numpy as np
from scipy.stats import iqr

happiness_Q1 = happiness_df.describe().T.loc['gdp_per_capita', '25%']
happiness_Q3 = happiness_df.describe().T.loc['gdp_per_capita', '75%']

print(happiness_Q1, happiness_Q3)

iqr_value = happiness_Q3 - happiness_Q1
print(iqr_value)

iqr_value = iqr(happiness_df.gdp_per_capita)
print(iqr_value)

lower_bound = happiness_Q1 - 1.5 * iqr_value
upper_bound = happiness_Q3 + 1.5 * iqr_value
print(f'정상치 범위: {lower_bound} ~ {upper_bound}')
```

```python
import numpy as np
from scipy.stats import iqr

happiness_Q1 = happiness_df.describe().T.loc['happiness_score', '25%']
happiness_Q3 = happiness_df.describe().T.loc['happiness_score', '75%']

print(happiness_Q1, happiness_Q3)

iqr_value = happiness_Q3 - happiness_Q1
print(iqr_value)

iqr_value = iqr(happiness_df.happiness_score)
print(iqr_value)

lower_bound = happiness_Q1 - 1.5 * iqr_value
upper_bound = happiness_Q3 + 1.5 * iqr_value
print(f'정상치 범위: {lower_bound} ~ {upper_bound}')
```

```python
import numpy as np
from scipy.stats import iqr

happiness_Q1 = happiness_df.describe().T.loc['social_support', '25%']
happiness_Q3 = happiness_df.describe().T.loc['social_support', '75%']

print(happiness_Q1, happiness_Q3)

iqr_value = happiness_Q3 - happiness_Q1
print(iqr_value)

iqr_value = iqr(happiness_df.social_support)
print(iqr_value)

lower_bound = happiness_Q1 - 1.5 * iqr_value
upper_bound = happiness_Q3 + 1.5 * iqr_value
print(f'정상치 범위: {lower_bound} ~ {upper_bound}')
```

```python
hp_mean_df = happiness_df.groupby('region').mean('gdp_per_capita').reset_index()
display(hp_mean_df)

hp_min_df = happiness_df.groupby('region').min('gdp_per_capita').reset_index()
display(hp_min_df)

hp_max_df = happiness_df.groupby('region').max('gdp_per_capita').reset_index()
display(hp_max_df)

hp_std_df = happiness_df.groupby('region')['gdp_per_capita'].std().reset_index()
display(hp_std_df)
```

```python
hp_mean_df_h = happiness_df.groupby('region').mean('happiness_score').reset_index()
display(hp_mean_df_h)

hp_min_df_h = happiness_df.groupby('region').min('happiness_score').reset_index()
display(hp_min_df_h)

hp_max_df_h = happiness_df.groupby('region').max('happiness_score').reset_index()
display(hp_max_df_h)

hp_std_df_h = happiness_df.groupby('region')['happiness_score'].std().reset_index()
display(hp_std_df_h)
```

```python
hp_mean_df_s = happiness_df.groupby('region').mean('social_support').reset_index()
display(hp_mean_df_s)

hp_min_df_s = happiness_df.groupby('region').min('social_support').reset_index()
display(hp_min_df_s)

hp_max_df_s = happiness_df.groupby('region').max('social_support').reset_index()
display(hp_max_df_s)

hp_std_df_s = happiness_df.groupby('region')['social_support'].std().reset_index()
display(hp_std_df_s)
```
